```python
import requests
import pandas as pd

def get_coordinates(address, api_key):
    """
    Fetches latitude and longitude for a given address using Google Geocoding API.

    :param address: Address to geocode.
    :param api_key: Google API key.
    :return: A tuple with (latitude, longitude).
    """
    base_url = "https://maps.googleapis.com/maps/api/geocode/json"
    params = {
        "address": address,
        "key": api_key
    }

    response = requests.get(base_url, params=params)
    if response.status_code == 200:
        data = response.json()
        if data['status'] == 'OK':
            # Extract latitude and longitude
            location = data['results'][0]['geometry']['location']
            return location['lat'], location['lng']
        else:
            raise ValueError(f"Geocoding failed with status: {data['status']}")
    else:
        response.raise_for_status()

def std_codes(code):
    if str(code).replace('.','').isdigit():
        return(str(int(code))).replace('_x000D_\n', '').replace('\n', '')
    else:
        return (str(code)
                .replace('_x000D_\n', '')
                .replace('\n', '')
                .replace('.',''))
    

API_KEY = 'AIzaSyB4aq_pvbkaeGga3SHTvG8nj5jBBAdJpH8'
```
